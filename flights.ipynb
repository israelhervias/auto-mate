```python
conn = sqlite3.connect("flight.sqlite3")
c = conn.cursor()
```

```python
create_table_flights = """
    CREATE OR REPLACE TABLE IF NOT EXITS flights (
        [index] TEXT,
        [flight_date] TEXT,
        [departure_airport] TEXT,
        [departure_terminal TEXT,
        [departure_gate] TEXT,
        [departure_delay] TEXT,
        [departure_scheduled] TEXT.
        [departure_estimated] TEXT,
        [arrival_airport] TEXT,
        [arrival_terminal TEXT,
        [arrival_gate] TEXT,
        [arrival_delay] TEXT,
        [arrival_scheduled] TEXT,
        [arrival_estimated] TEXT,
        [airline_name] TEXT,
        [flight_number] TEXT
    );
"""

c.execute(create_table_flights)
conn.commit()
```

```python
create_flights_query = """
    CREATE TABLE IF NOT EXISTS flight (
        [index] TEXT,
        [flight_date] TEXT,
        [departure_airport] TEXT,
        [departure_terminal] TEXT,
        [departure_gate] TEXT,
        [departure_delay] REAL,
        [departure_scheduled] TEXT,
        [departure_estimated] TEXT,
        [arrival_airport] TEXT,
        [arrival_terminal] TEXT,
        [arrival_gate] TEXT,
        [arrival_delay] REAL,
        [arrival_scheduled] TEXT,
        [arrival_estimated] TEXT,
        [airline_name] TEXT,
        [flight_number] TEXT        
    );
"""
c.execute(create_flights_query)
conn.commit()
```
